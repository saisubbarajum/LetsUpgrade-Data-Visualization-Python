```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fmri = sns.load_dataset("fmri")

fmri.head()
```

**line** **plot** 

```python
sns.lineplot(x = "timepoint", y = "signal", data = fmri)
```

**bar graph**

```python
sns.barplot(x = "timepoint", y = "signal", data = fmri)
```

**rel plots**

```python
fmri.describe()
```

```python
sns.relplot(x = "timepoint", y = "signal",kind='line', data = fmri)
```

```python
sns.relplot(x = "timepoint", y = "signal", data = fmri, hue='signal', style='timepoint')
```

**cat plots**

```python
sns.catplot(x='timepoint', y = "signal", data = fmri)
```

```python
sns.catplot(x='timepoint', y = "signal", kind='box', data = fmri)
```

```python
sns.catplot(x='timepoint', y = "signal", kind='boxen', data = fmri)
```

```python
sns.catplot(x='timepoint', y = "signal", kind='swarm', data = fmri)
```
