```python
days = [1,2,3,4,5,6,7] #days of the week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2

plt.plot(days,sales_1)
plt.plot(days,sales_2)
plt.title("Days vs Sales")
plt.xlabel("Days")
plt.ylabel("Sales of 1 and 2")
```

```python
x = [1,2,3,4]
y1 = [20,30,40,50]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]

plt.subplot(3,3,1)
plt.plot(x,y1)
plt.subplot(3,3,2)
plt.plot(x,y2,"g*--")
plt.subplot(3,3,3)
plt.plot(x,y3,"ro--")
plt.subplot(3,3,4)
plt.plot(x,y4,"b--")
plt.subplot(3,3,5)
plt.plot(x,y5,"y--")
```
